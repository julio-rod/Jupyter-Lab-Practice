## Step 1: Import the Pandas and `pathlib` libraries.

```python
# Import the Pandas library
import pandas as pd


# Import the Path module from the pathlib library
from pathlib import Path
```

## Step 2: Use `Path` with the `read_csv` function to read the CSV file into the DataFrame. Review the first five rows of the DataFrame to confirm the import.

```python
# Read in the CSV file called "country_bond_rates.csv" using the Path module
# The CSV file is located in the Resources folder
# Set the index to the column "Date"
# Set the parse_dates and infer_datetime_format parameters
bond_rates_df = pd.read_csv(
    Path("../Resources/country_bond_rates.csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True)
# Review the first 5 rows of the DataFrame
bond_rates_df.head()
```

## Step 3: Use the `iloc` function to select the first row of the bond yield data.

```python
# Use the iloc function to select the first row of the Dataframe 
bond_rates_df.iloc[0]
```

## Step 4: Use the `iloc` function to instead return the first 10 rows of data.

```python
# Use the iloc function to select the first ten rows of the Dataframe 
bond_rates_df.iloc[0:10]
```

## Step 5: Use the `loc` function to select the full year of data for 2019. Generate summary statistics for this data by using the `describe` function.

```python
# Use the loc function to select data from the full year of 2019 
bond_rates_2019_df = bond_rates_df.loc["2019-01-01":"2019-12-31"]

# Use the describe function to generate the summary statistics for the `bond_rates_2019_df` DataFrame
bond_rates_2019_df.describe()
```

## Step 6: Use the `loc` function to select all the data from 2020. Generate summary statistics for this data by using the `describe` function.

```python
# Use the loc function to select data from the full year of 2020 
bond_rates_2020_df = bond_rates_df.loc["2020"]

# Use the describe function to generate the summary statistics for the `bond_rates_2020_df` DataFrame
bond_rates_2020_df.describe()
```
