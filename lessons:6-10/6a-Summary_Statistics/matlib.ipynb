```python
import pandas as pd
import numpy as np
daily_returns = pd.DataFrame(
    np.random.randn(10, 2),
    index=pd.date_range('1/1/2020', periods=10),
    columns=["AAPL", "GOOG"])
daily_returns.describe()
```

```python
daily_returns.plot() #using .plot
```

```python
# With the kind parameter, you can specify the type of plot that you want. The following list has the values you use for some of the most common types of plots:

# line: This is the default Pandas plot. Its great for data that involves dates and times (that is, time series data).
# bar: This plot is best when the dataset contains labeled categories of data, like age group or income range.
# box: This plot is best for visualizing the distribution of values within each column.
# hist: Short for histogram, this visualization displays the distribution of data by illustrating how often each value (or group of values) in a set of data occurs.
# (That is, it illustrates the frequency of each value or group of values.)
# scatter: This plot is best when you have numeric columns for x and y values.
# -Spot
```

```python
daily_returns.plot(kind='box') #using `kind` parameter
```

```python
# The box plot for Apple has a long upper whisker, which indicates that the values in the fourth quartile of data points are quite spread out. 
# (The fourth quartile consists of the values from 75% to max in the summary statistics.) The Google box, by contrast, has a long lower whisker. 
# This illustrates that the first quartile of data points is quite spread out. (The first quartile consists of the values from min to 25% in the summary statistics.) 
# These whiskers help show why the mean value for Apples performance is higher than that for Googles, based on the values in the DataFrame.
# -spot
```
